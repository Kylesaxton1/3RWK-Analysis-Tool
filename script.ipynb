# Three Rivers Waterkeeper - Script

```python
import requests
import pandas as pd
import numpy as np

baseURL = "https://echodata.epa.gov/echo/eff_rest_services.get_effluent_chart"
jsonOut = "&output=JSON"
```

```python
permitNumber = input("Enter permit number: ")
#?p_id = PAR606135
```

```python
currURL = baseURL + "?p_id=" + permitNumber + jsonOut
response = requests.get(currURL)
response
```

```python
data = response.json()
```

```python
facility = data['Results']
facility_df = pd.DataFrame(data["Results"])
singlefacility_df = pd.DataFrame(facility_df.iloc[0])
singlefacility_df
```

```python
permitFeatures = facility['PermFeatures']
permfeat_df = pd.DataFrame(permitFeatures)
showpermfeat_df = pd.DataFrame(permfeat_df.iloc[0])
showpermfeat_df
```

```python
params = pd.DataFrame(permfeat_df.loc[0]['Parameters'])
params
```

```python
oxygendmr_df = pd.DataFrame(params.loc[0]['DischargeMonitoringReports'])
oxygendmr_df                  
```
